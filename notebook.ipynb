Classification results (First Extraction)

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

classification_results_path = "classification_results_first.csv"
classification_results = pd.read_csv(classification_results_path)

correct = classification_results[classification_results['label'] == classification_results['classification']]
accuracy = len(correct) / len(classification_results)
print(f"Accuracy: {accuracy:.2%}")


labels = classification_results['label'].unique()

precision = precision_score(classification_results['label'], classification_results['classification'], average=None, labels=labels)
recall = recall_score(classification_results['label'], classification_results['classification'], average=None, labels=labels)
f1 = f1_score(classification_results['label'], classification_results['classification'], average=None, labels=labels)

print("\nMetrics by class:")
for i, label in enumerate(labels):
    print(f"{label}: Precision={precision[i]:.2f}, Recall={recall[i]:.2f}, F1={f1[i]:.2f}")

weighted_precision = precision_score(classification_results['label'], classification_results['classification'], average='weighted')
weighted_recall = recall_score(classification_results['label'], classification_results['classification'], average='weighted')
weighted_f1 = f1_score(classification_results['label'], classification_results['classification'], average='weighted')

print(f"\nWeighted Averages: Precision={weighted_precision:.2f}, Recall={weighted_recall:.2f}, F1={weighted_f1:.2f}")

cm = confusion_matrix(classification_results['label'], classification_results['classification'], labels=labels)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
